```python
import pandas as pd

df_tzmap = pd.read_csv("iata_tzmap.csv", sep="\t", header=None, names=["iata_code", "tz"])
df = pd.read_csv("optd_por_public.csv", sep="^")

df
```

```python
df = df.join(df_tzmap.set_index("iata_code"), on="iata_code", how="inner").drop_duplicates(subset="iata_code").reset_index()
df = df[["iata_code", "name", "tz"]]
df
```

```python
# Convert df into a .ts file for a list of objects, keys in object are the iata, name and tz
# Escape double quotes in the name

with open("iata_tzmap.ts", "w") as f:
    f.write("export const iata_tzmap = [\n")
    for i, row in df.iterrows():
        f.write(
            "  {"
            + 'iata: "{}", name: "{}", tz: "{}"'.format(
                row["iata_code"], row["name"].replace('"', '\\"'), row["tz"]
            )
            + "}"
        )
        if i < len(df) - 1:
            f.write(",")
        f.write("\n")
    f.write("];\n")
```
